# Scikit-Learn Workflow
Scikit-learn provides a streamlined workflow for building and evaluating machine learning models. The typical workflow includes:

1. **Data Loading**: Import datasets using pandas or scikit-learn's built-in loaders.
2. **Data Exploration**: Analyze and visualize the data to understand distributions, missing values, and relationships.
3. **Data Preprocessing**: Handle missing values, encode categorical variables, and scale features as needed.
4. **Feature Engineering**: Create new features or transform existing ones to improve model performance.
5. **Model Selection**: Choose an appropriate algorithm for the problem (e.g., classification, regression).
6. **Model Training**: Fit the model to the training data.
7. **Model Evaluation**: Assess model performance using metrics and validation techniques.
8. **Hyperparameter Tuning**: Optimize model parameters for better results.
9. **Prediction**: Use the trained model to make predictions on new data.
10. **Deployment**: Integrate the model into production or further analysis.

This notebook demonstrates each step using real datasets and scikit-learn tools.

## Value Imputation

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'mean')
```

```python
df['age'] = imputer.fit_transform(df[['age']])
```

```python
df.age.isnull().sum()
```

```python
def get_parameters(df):
    parameters = {}
    for col in df.columns[df.isnull().any()]:
        if df[col].dtype == 'float64' or df[col].dtype == 'int32':
            strategy = 'mean'
        else:
            strategy = 'most_frequent'

        missing_values = df[col][df[col].isnull()].values[0]
        parameters[col] = {'missing_values': missing_values,'strategy': strategy}
    return parameters

parameters = get_parameters(df)
```

```python
parameters = get_parameters(df)
parameters
```

```python
for col, param in parameters.items():
    missing_values = param['missing_values']
    strategy = param['strategy']
    imputer = SimpleImputer(missing_values=missing_values, strategy=strategy)
    df[col] = imputer.fit_transform(df[[col]]).ravel()
```
